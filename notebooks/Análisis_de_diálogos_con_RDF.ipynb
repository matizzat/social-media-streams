```python
from rdflib import Graph

# Create a Graph, parse in Internet data
g = Graph().parse("http://www.w3.org/People/Berners-Lee/card")

# Query the data in g using SPARQL
# This query returns the 'name' of all `foaf:Person` instances
q = """
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>

    SELECT ?name
    WHERE {
        ?p rdf:type foaf:Person .

        ?p foaf:name ?name .
    }
"""

# Apply the query to the graph and iterate through results
for r in g.query(q):

  print(r["name"])

# prints: Timothy Berners-Lee
```

```python
from rdflib import Graph, Literal, Namespace, RDF

# Initialize graph and namespaces
g = Graph()
EX = Namespace("https://example.org")

# Define your custom entities
kaka = EX["person/kaka"]
kiki = EX["person/kiki"]
has_friend = EX["ontology/hasShit"]
name_property = EX["ontology/name"]

# Add triples to the graph
g.add((kaka, RDF.type, EX["ontology/Person"]))
g.add((kaka, name_property, Literal("Kaka")))
g.add((kaka, has_friend, kiki))

# Serialize to see the result
print(g.serialize(format="ntriples"))
```

```python
from pyvis.network import Network
from pyvis import network as net
import networkx as nx

def RDFgraph_to_network(G_rdf):
  G = nx.MultiDiGraph()

  for s, r, t in G_rdf:

    if(s not in G):
      G.add_node(s, label=s, color = 'black')

    if(t not in G):
        G.add_node(t, label=t, color = 'black')

    color = 'blue'
    width = 1

    G.add_edge(
        s,
        t,
        label=r,
        title = r,
        color = color,
        width = width
    )

  return G

G = RDFgraph_to_network(G_rdf)
vis = net.Network(directed=True)
vis.from_nx(G)
# Set the height and width of the network:
vis.height = "500px"
vis.width = "100%"
vis.cdn_resources = 'remote'

# Prepare the notebook resources to initialize the template engine
vis.prep_notebook()

# Show the network:
vis.show('network.html', notebook=True)
```
