Para ejecutar la notebook es un prerrequisito subir al entorno de ejecución de Google Colab el lote de datos de la Tarea 2 de la eRisk 2025, titulada «Detección anticipada de depresión contextualizada».

```python
from typing import List, Dict
import json
import os

dataset_path: str = "./dataset"

class Server:

  _round: int = 0

  def get(self) -> List[Dict]:

    global dataset_path

    result = []
    for filename in os.listdir(dataset_path):

      with open(dataset_path + '/' + filename, 'r') as f:
        submissions = json.load(f)

      if(self._round < len(submissions)):
        result.append(submissions[self._round])

    self._round += 1
    return result
```
